```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pathlib
import time
```

```python
random_df = pd.read_csv('random_points_100.csv')
random_df
```

```python
dpi = np.arange(10,1000)
dpi
```

```python
def create_plot(ax, dpi, x, y, directory):
    ax.plot(x,y)
    
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('test title')
    
    plt.savefig(directory / f'{dpi}.png')
```

```python
plot_dir = pathlib.Path('plots') / '100'
```

```python
fig, ax = plt.subplots()

for d in dpi:
    fig.dpi = d
    
    create_plot(ax, d, random_df['x'], random_df['y'], plot_dir)
    
    fig.clear()
```
